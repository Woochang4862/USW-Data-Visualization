## 실습4

```python
## 코드작성
def foo1(v1, v2,op):
    if op=='+':
        res=(v1+v2)
    if op=='-':
        res=(v1-v2)
    if op=='*':
        res=(v1*v2)
    if op=='/':
        res=(v1/v2)
    return res
## 코드작성 끝

foo1(3,2,'+')
foo1(1,5,'-')
foo1(2,3,'*')
foo1(4,8,'/')
```

## 실습5

```python
# 단어가 들어있는 list에서 숫자가 2개이상 포함되어있는 단어 출력하는 프로그램이다.
lst1=['12ab','abc1','abc123','abc12','aaaaacc','11']

for word in lst1:
    ## 코드작성
    num_cnt=0
    for ch in word:
        if ch.isdigit():
            num_cnt+=1
            
    if num_cnt>=2:
        print(word)
    ## 코드작성 끝
```
