# plot (기본)

## plot 예제

```python
N = 5
fig=plt.figure(figsize=(N,4), dpi=100)
ax=fig.subplots()


X=range(1,N+1)
Y=[0 if i % 2 == 0 else 1 for i in X]

ax.plot(X,Y)
```

```python
fig=plt.figure(figsize=(4,4), dpi=100)
ax=fig.subplots()

X=range(100)
Y=[value**2 for value in X]

list(X[:5])
list(Y[:5])

ax.plot(X,Y)
```

# plot with format

```python
fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

X=np.arange(0,100,10) # numpy arange
Y=X**2

ax.plot(X, Y, '_b') # default ',-b'
```

```python
fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

X=np.arange(10)
Y=(X-5)**2

ax.plot(X,Y, '.--r')
```
