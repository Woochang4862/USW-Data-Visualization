## 문제9

```python
# 'data/data_score.txt'를 읽은 data가 있다
# mid column을 X로, final column을 Y로 하여 scatter 하려고 한다. 아래의 조건에 맞게 scatter 하시오

# 조건1) 여학생(gender:F)은 '>' marker, 남학생(gender:M)은 's' marker
# 조건2) att 8점이상: color='b', att 8점 미만: color='y'
# 조건3) scatter 함수를 두 번 만 호출하시오

data = pd.read_table('data/data_score.txt', sep='\t')
data.head()

fig=plt.figure(figsize=(5,3), dpi=100)
ax=fig.subplots()

## 코드작성
for gender, marker in zip(['M','F'],['s','>']):
    condition = data['gender']== gender
    colors = np.where(data.loc[condition,'att']>=8,'b','y')
    _=ax.scatter(data.loc[condition,'mid'], data.loc[condition,'final'], color=colors, marker=marker)
```

## 문제10

```python
# 중심극한정리는 
# 모집단의 분포가 무엇이든 관계없이 n이 크면 표본평균의 분포가 근사적으로 정규분포를 따르고,
# 그 정규분포의 평균은 모집단의 평균과 같고, 표준편차는 모집단의 표준편차/sqrt(n) 이다 (PPT 참고))

# 아래 특정 모분포에서 여러개의 표본들을 구하고 그들의 평균들을 가지고 그린 히스토그램이 있다.
# 이 히스토그램에 더하여, 이 히스토그램에 fitting하는 정규분포의 pdf 함수를 그리시오.

# 힌트1: 모집단의 평균과 표준편차는 모집단에서 1000개의 값을 샘플링하여 평균(np.mean)과 표준편차(np.std)를 구하여 사용
# 주의1: fitting이 정확하게 안될수도 있음. 중심극한정리에 따른 pdf 함수를 그리면 정답 인정

fig=plt.figure(figsize=(5,3), dpi=100)
ax=fig.subplots()

data=np.random.chisquare(4,size=(3000,1000))
m1=data.mean(axis=1)

_=ax.hist(m1, bins=20, edgecolor='k', density=True)

## 코드작성
def pdf(X,mu,sigma):
    a = 1/(sigma*np.sqrt(2*np.pi))
    b = -(X-mu)**2/(2*sigma**2)
    return a*np.exp(b)

X = np.linspace(3,5,1000)
mu = data.mean()
sigma = data.std()/np.sqrt(1000)
_=ax.plot(X,pdf(X,mu,sigma), 'r-')
mu, sigma
```

```python
# 'data/data_score.txt'를 읽은 data가 있다
# 각 항목 (mid, final, att, proj) 별로,
# 1학년(year: 1)의 평균 점수와 4학년(year: 4)의 평균 점수를 구하고, 이들을 다중 bar그래프로 나타내시오 (1학년, 4학년 순)

# 주의1: 각 항목 별 bar들이 떨어질수 있도록,  bar 그룹간의 간격을 어느정도 둬야함 (간격의 크기는 상관없음).
# 주의2: axes bar 함수를 사용할 것.

data = pd.read_table('data/data_score.txt', sep='\t').drop(columns=['name', 'gender'])
data.head()

fig=plt.figure(figsize=(5,4), dpi=100)
ax=fig.subplots()

## 코드작성
mean1 = data[data['year']==1].drop(columns=['year']).mean(axis=0)
mean4 = data[data['year']==4].drop(columns=['year']).mean(axis=0)

X = np.arange(len(mean1.index))
w = 0.3
_=ax.bar(X,mean1,width=w)
_=ax.bar(X+w,mean4,width=w)
_=ax.set_xticks(X+0.15, labels=data.columns[1:])
```

## 문제12

```python
# 주어진 X,Y를 ax1에 scatter 하였다.
# 아래의 조건을 만족시키면서 주어진 X,Y를 ax2에 scatter 하시오
# - 0<=Y<1 인 경우: color: color_list[0], marker: marker_list[0]
# - 1<=Y<2 인 경우: color: color_list[1], marker: marker_list[1]
# - 2<=Y<3 인 경우: color: color_list[2], marker: marker_list[2]

# 주의1: color_list, marker_list의 값은 채점시 변경 가능

# 힌트1 (사용하지 않아도 상관없음):
 # a=0.12
 # int(a) ==> 0
 # b=1.42
 # int(b) ==> 1

fig=plt.figure(figsize=(10,4), dpi=100)
ax1, ax2 =fig.subplots(1,2)

data=np.random.uniform(0,3,size=(100,2))
X=data[:,0]
Y=data[:,1]

## ax1
_=ax1.scatter(X, Y, edgecolor='k', s=50)

## ax2
marker_list=['>','^','<']
color_list=['r','g','b']

## 코드작성
for i in range(len(data)):
    _=ax2.scatter(X[i],Y[i], marker=marker_list[int(Y[i])], color=color_list[int(Y[i])], edgecolor='k')
```
