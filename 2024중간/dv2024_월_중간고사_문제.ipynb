## 문제2

```python
# -4<X<3 범위에서 Y=(X+3)*(X+1)*(X-2)를 plot 하고, Y의 기울기 함수 Y1을 plot하였다.
# Y의 2차 기울기 함수 Y2 (즉, Y1의 기울기 함수)를 추가로 plot하시오

# 주의1) 원래 강의자료의 plot_slope 함수에 return이 추가됨

def plot_slope(X1,Y1,ax1):
    Xs=X1[1:]-X1[:-1]
    Ys=Y1[1:]-Y1[:-1]
    _=ax1.plot(X1[1:], Ys/Xs, 'b')
    
    return Ys/Xs
    
    
fig=plt.figure(figsize=(4,3), dpi=100)
ax=fig.subplots()

X=np.linspace(-4,3,300)
Y=(X+3)*(X+1)*(X-2)

_=ax.plot(X,Y, color='r')
Y1=plot_slope(X, Y , ax)

## 코드작성
_=plot_slope(X[1:],Y1,ax)

ax.axvline(x=-2.12, color='gray', linestyle=':')
ax.axvline(x=0.79, color='gray', linestyle=':')
ax.axhline(y=0, color='gray', linestyle=':')
ax.axvline(x=-0.67, color='k', linestyle='-')
```

## 문제9

```python
# 'data/data_score.txt'를 읽은 data가 있다
# mid column을 X로, final column을 Y로 하여 scatter 하려고 한다. 아래의 조건에 맞게 scatter 하시오

# 조건1) 해당 학생의 proj 점수를 투명도에 반영하시오 (proj 점수가 높을수록 진하게)
# 조건2) 마커 사이즈: 100

fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

data = pd.read_table('data/data_score.txt', sep='\t').drop(columns=['name', 'gender'])
data.head()

## 코드작성
mx = data['proj'].max()
mn = data['proj'].min()
alpha_list = (data['proj'] - mn + 1) / (mx - mn + 1)
_=ax.scatter(data['mid'], data['final'], alpha=alpha_list, s=100)
```

## 문제10

```python
# 'data/data_stock.txt'를 읽은 data가 있다
# 이 data는 4개 종목의 9시~15시까지의 주식 가격 이다.
# 아래의 두가지 조건을 만족하며 plot하시오

# 조건1) 각 종목별로 해당 linestyle과 marker로 line 그래프 plot (사전에 정의되어 있음)
# 조건2) 이날 가격이 하락한 종목은 파란색, 상승한 종목은 빨간색으로 plot (9시 가격과 15시 가격을 비교)


data = pd.read_table('data/data_stock.txt', sep='\t', index_col=0).T
data

marker_dic={'p1':'o','p2':'+','p3':'s','p4':'>'}
line_dic ={'p1':':','p2':'-','p3':'--','p4':'-.'}

fig=plt.figure(figsize=(6,3), dpi=100)
ax=fig.subplots()

## 코드작성
for col in data.columns:
    color = 'r' if data[col][0] < data[col][-1] else 'b'
    _=ax.plot(data.index, data[col], marker=marker_dic[col], linestyle=line_dic[col], color=color)

```

## 문제11

```python
# data의 X, Y column을 scatter 하는데,
# X-1<Y<X+1인 경우는  marker 's', color 'y',
# 그렇지 않은 경우는   marker 'o', color 'b' 로 하시오


fig=plt.figure(figsize=(6,5), dpi=100)
ax=fig.subplots()

data=pd.DataFrame(np.random.uniform(-3,3, size=(500,2)), columns=['X','Y'])
data

## 코드작성
X = data['X']
Y = data['Y']
for marker, color, condition in zip(['s','o'], ['y','b'], [np.bitwise_and(X-1<Y, Y<X+1), np.bitwise_or(X-1>=Y, Y>=X+1)]):
    _data = data[condition]
    _=ax.scatter(_data['X'],_data['Y'],marker=marker, color=color, edgecolor='k')
```

## 문제12

```python
# 'data/data_class.txt'를 읽은 data가 있다 (10개의 feature, class는 사용하지 않음).
# 표준편차가 가장 작은 feature와 가장 큰 feature의 데이터를 boxplot으로 그리시오

# 조건1) 표준편차가 작은 feature, 큰 feature 순으로 그리시오
# 조건2) 해당 feature를 'selected_feature' 변수에 담으시오

fig=plt.figure(figsize=(5,3), dpi=100)
ax=fig.subplots()

## data load
data = pd.read_table('data/data_class.txt', sep='\t').drop(columns='class')

## 코드작성
std = data.std(axis=0)
mx = std.argmax()
mn = std.argmin()
selected_feature = [std.index[mn], std.index[mx]]
print(selected_feature)
_=ax.boxplot([data[selected_feature[0]],data[selected_feature[1]]])
```
