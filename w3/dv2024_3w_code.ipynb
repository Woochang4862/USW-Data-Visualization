# plot (곡선 그리기)

## sin 곡선 그리기

```python
import math

fig=plt.figure(figsize=(6,6), dpi=100)
(ax1, ax2)=fig.subplots(2,1)

## without numpy
T=range(100)
X=[(2*math.pi*t)/len(T) for t in T]
Y=[math.sin(value) for value in X]

ax1.plot(X,Y)
ax1.set_title("without numpy")

## with numpy
X=np.linspace(0,2*np.pi, 100)
Y=np.sin(X)

ax2.plot(X,Y)
ax2.set_title("with numpy")
```

## 실습1

```python
fig=plt.figure(figsize=(7,8), dpi=100)
(ax1, ax2)=fig.subplots(2,1)

## 코드작성
ax1.set_title('sin(X-pi)')
X = np.linspace(0,2*np.pi,100)
Y = np.sin(X-np.pi)
ax1.plot(X,Y,'.:g')

ax2.set_title('X^2-2X+1')
X = np.linspace(-4,6,100)
Y = X**2-2*X+1
ax2.plot(X,Y)
```

# 다중 곡선 그리기

```python
fig=plt.figure(figsize=(7,4), dpi=100)
ax=fig.subplots()

X=np.linspace(0,2*np.pi,100)
Ya=np.sin(X)
Yb=np.cos(X)

ax.plot(X,Ya)
ax.plot(X,Yb)
ax.set_title("sin & cos")
```

## 기울기 함수 그리기

```python
def plot_slope(X1,Y1,ax1):
    Xs=X1[1:]-X1[:-1]
    Ys=Y1[1:]-Y1[:-1]
    ax1.plot(X1[1:], Ys/Xs, 'b')
    
fig=plt.figure(figsize=(7,4), dpi=100)
ax=fig.subplots()

X=np.linspace(-3,3,100)
Y=np.exp(-X**2)

ax.plot(X,Y,'r')
plot_slope(X,Y,ax)
```

## 실습5

```python
fig=plt.figure(figsize=(13,4), dpi=100)
(ax1,ax2,ax3,ax4) =fig.subplots(1,4)

data=pd.read_table('data/3w_d3.txt', sep='\t')

type(data)
data.shape
data.head()

X=range(len(data))

# ax1 (axe plot, for loop)
for col in data.columns:
    _=ax1.plot(X,data[col])

# ax2 (pandas plot)
data.plot.line(ax=ax2)

# ax3 (각 column 평균이 50 이상 80이하인 컬럼은 빨간색, 아닌 컬림은 파란색)
for col in data.columns:
    m1=data[col].mean()
    if (50<=m1) and (m1<=80):
        _=ax3.plot(X,data[col],'r-')
    else:
        _=ax3.plot(X,data[col],'b-')

# # ax3 (pandas plot)
# color_list=data.mean(axis=0).map(lambda m1: 'r' if (m1>=50) and (m1<=80) else 'b')
# _=data.plot.line(ax=ax3, color=color_list)

# ax4 (특정 column set['Y1','Y4','Y5','Y9'] 에 대해서 다른 색깔로 plot하기)
for col in data.columns:
    if col in ['Y1','Y4','Y5','Y9']:
        _=ax4.plot(X,data[col],'r-')
    else:
        _=ax4.plot(X,data[col],'b-')
```

## 실습7

```python
fig=plt.figure(figsize=(10,4), dpi=100)
ax1, ax2 =fig.subplots(1,2)

data=np.random.uniform(-1.5,1.5,size=(1000,2))

X=data[:,0]
Y=data[:,1]

_=ax1.scatter(X, Y, edgecolor='k')

## 코드작성
blue_condition = data[:,0] <= data[:,1]
blue_case = data[blue_condition]
red_case = data[np.invert(blue_condition)]
_=ax2.scatter(blue_case[:,0], blue_case[:,1], c='b')
_=ax2.scatter(red_case[:,0], red_case[:,1], c='r')
```

## 특정 함수에 noise 추가해서 scatter 해보기
