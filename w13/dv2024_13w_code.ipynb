```python
# df1의 x,y를 빨간색으로, df2의 x,y를 파란색으로 scatter 하였다.
# df1의 점 중 가장 오른쪽에 있는 점을 p1이라고하고, df2의 점 중 가장 왼쪽에 있는 점을 p2라고 할때,
# 1) p1, p2에 추가로 노란 별표를 scatter 하고
# 2) p1,p2를 화살표로 연결하시오

# - A: marker='*', color='r'
# - B: marker='o', color='b'

fig=plt.figure(figsize=(7,3), dpi=100)
ax1=fig.subplots()

df1=pd.DataFrame({'x':np.random.normal(-3,0.5,size=50),
                  'y':np.random.uniform(-1,1,size=50)})
df1.head()

df2=pd.DataFrame({'x':np.random.normal(3,0.5,size=50),
                  'y':np.random.uniform(-1,1,size=50)})
df2.head()

_=ax1.scatter(df1['x'],df1['y'], color='r', edgecolor='k', s=70)
_=ax1.scatter(df2['x'],df2['y'], color='b', edgecolor='k', s=70)

#=== 코드작성 ===#
right_point = df1.loc[df1['x'].argmax()]
left_point = df2.loc[df2['x'].argmin()]
_=ax1.scatter(*zip(right_point, left_point), marker='*', color='y')
_=ax1.annotate('',xy=right_point,xytext=left_point,arrowprops={'arrowstyle':'<|-|>'})
```

```python
## data load
df = pd.read_csv('data/seoul_starbucks.txt', sep='|')

## fig, axs creation
fig=plt.figure(figsize=(8, 5), dpi=150)
ax1, ax2=fig.subplots(1,2)

## ax1
count = df['시군구명'].value_counts().iloc[::-1]
_=ax1.barh(count.index,count)
for i, idx in enumerate(count.index):
    _=ax1.text(count.loc[idx],i,str(count.loc[idx]),ha='left',va='center')
_=ax1.set_xlim(0,90)

## ax2
count = count.sort_index(ascending=False)
mx = count.max()
mn = count.min()
## alpha sequence 안됨
# alphas = ((count - mn + 1)/(mx-mn+1)).values
# alphas
# _=ax2.barh(count.index,count,alpha=alphas,color='b')
for idx in count.index:
    alpha = (count.loc[idx]-mn+1)/(mx-mn+1)
    _=ax2.barh(idx,count.loc[idx],alpha=alpha,color='b')
fig.tight_layout()
```

```python
# df1의 x,y를 빨간색으로, df2의 x,y를 파란색으로 scatter 하였다.
# df1의 점 중 가장 오른쪽에 있는 점을 p1이라고하고, df2의 점 중 가장 왼쪽에 있는 점을 p2라고 할때,
# 1) p1, p2에 추가로 노란 별표를 scatter 하고
# 2) p1,p2를 화살표로 연결하시오

# - A: marker='*', color='r'
# - B: marker='o', color='b'

fig=plt.figure(figsize=(7,3), dpi=100)
ax1=fig.subplots()

df1=pd.DataFrame({'x':np.random.normal(-3,0.5,size=50),
                  'y':np.random.uniform(-1,1,size=50)})
df1.head()

df2=pd.DataFrame({'x':np.random.normal(3,0.5,size=50),
                  'y':np.random.uniform(-1,1,size=50)})
df2.head()

_=ax1.scatter(df1['x'],df1['y'], color='r', edgecolor='k', s=70)
_=ax1.scatter(df2['x'],df2['y'], color='b', edgecolor='k', s=70)

#=== 코드작성 ===#
mx,my=df1.sort_values(by='x').iloc[-1]
nx,ny=df2.sort_values(by='x').iloc[0]

_=ax1.scatter([mx,nx],[my,ny], marker='*', color='y')
_=ax1.annotate('',(mx,my), (nx,ny), arrowprops={'arrowstyle': '<->'})
#=== 코드작성 끝 ===#
```

### 실습7 (구별 매장의 개수를 bar 그래프로 나타내기)

```python
## data load
df = pd.read_csv('data/seoul_starbucks.txt', sep='|')

## fig, axs creation
fig=plt.figure(figsize=(8, 5), dpi=150)
ax1, ax2=fig.subplots(1,2)

## ax1
cnt1=df['시군구명'].value_counts()[::-1]
_=ax1.barh(cnt1.index, cnt1)

for ii, val in enumerate(cnt1):
    _=ax1.text(val, ii, val, va='center')

_=ax1.set_xlim(0,90)

## ax2
cnt1=df['시군구명'].value_counts()[::-1]
cnt2=cnt1.sort_index(ascending=False)

m1=cnt2.max()
n1=cnt2.min()

for ind in cnt2.index:
    _=ax2.barh(ind, cnt2.loc[ind], alpha=(cnt2.loc[ind]-n1+1)/(m1-n1+1), color='b')

fig.tight_layout()
```
