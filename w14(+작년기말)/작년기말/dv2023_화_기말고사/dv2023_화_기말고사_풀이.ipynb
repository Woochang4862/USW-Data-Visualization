```python
# title_list에 각 axes에 들어갈 title이 담겨져 있다.
# 결과처럼 title이 들어가도록 코드를 완성하시오

# 조건1) for loop 사용
# 조건2) title_list의 값은 채점 시 변경 가능

fig=plt.figure(figsize=(5,3), dpi=100)
axs=fig.subplots(2,3).flatten()

title_list=['a', 'c', 'e', 'd', 'f', 'b']

#=== 코드작성 ===#
for ax, title in zip(axs, title_list):
    _=ax.set_title(title)
#=== 코드작성 끝===

fig.tight_layout()
```

## 문제5

```python
# 주어진 데이터를 log2 scale로 아래와 같이 bar 그래프를 그렸다.
# 원래 값을 bar 위쪽에 text로 아래 그림과 같이 넣으시오

# 조건1) bar의 한 가운데 위치하도록 하시오
# 조건2) text가 bar에 닿지 않도록 하시오

fig=plt.figure(figsize=(8,3), dpi=100)
ax1=fig.subplots()

data=np.array([32,8,16,4,2])
X=range(len(data))

#=== 코드작성 ===#
data_log=np.log2(data)
_=ax1.bar(X,data_log)

for x, d, dlog in zip(X, data, data_log):
    _=ax1.text(x,dlog,d, ha='center', va='bottom', fontsize=16)
#=== 코드작성 끝 ===#

_=ax1.set_ylim(0,7)
_=ax1.grid(axis='y')
```

## 문제9

```python
# ax1에 df1에 대한 heatmap을 그렸다.
# ax2에도 df1에 대한 heatmap을 그리는데,
# 각 셀에 실제 값을 나타내는 대신, Even (짝수), Odd (홀수)를 나타내시오

fig=plt.figure(figsize=(12,4), dpi=100)
ax1,ax2=fig.subplots(1,2)

df1=pd.DataFrame({'A':[1,2,3], 'B':[2,3,4],'C':[3,4,5],'D':[4,5,6]}, index=['R1','R2','R3'])
df1

## ax1
hm1=ax1.pcolor(df1, cmap='GnBu', edgecolor='k')
_=fig.colorbar(hm1, ax=ax1)

col_len=len(df1.columns)
row_len=len(df1.index)
for r in range(row_len):
    for c in range(col_len):
        _=text = ax1.text(c+0.5, r+0.5, df1.iloc[r, c],ha="center", va="center", color="k", fontsize=16)
        
_=ax1.set_xticks(np.arange(col_len)+0.5, df1.columns)
_=ax1.set_yticks(np.arange(row_len)+0.5, df1.index)

## ax2
## 코드작성
hm1=ax2.pcolor(df1, cmap='GnBu', edgecolor='k')
_=fig.colorbar(hm1, ax=ax2)

col_len=len(df1.columns)
row_len=len(df1.index)
for r in range(row_len):
    for c in range(col_len):
        if df1.iloc[r, c]%2==0:
            _=text = ax2.text(c+0.5, r+0.5, 'Even',ha="center", va="center", color="k", fontsize=16)
        else:
            _=text = ax2.text(c+0.5, r+0.5, 'Odd' ,ha="center", va="center", color="k", fontsize=16)
_=ax2.set_xticks(np.arange(col_len)+0.5, df1.columns)
_=ax2.set_yticks(np.arange(row_len)+0.5, df1.index)
## 코드작성 끝
```

## 문제10

```python
# df는 x, y, marker column을 가지고 있다.
# marker에는 ['o','^','*'] 이 세 개의 값 중 하나가 들어있다.
# df의 (x,y)를 scatter 하는데, 해당 마커의 모양 (marker)으로 scatter 하시오

# 조건1) 마커의 색깔은 color_dic을 활용하시오 (key: marker)
# 조건2) legend에는 해당 마커의 색깔을 표시하시오
# 조건3) legend는 axes위에 한 행으로 나타내시오

fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

df = pd.read_table('data/tue_p11.txt', sep='\t').drop(columns='price')
df.head()

color_dic={'o': 'teal', '^': 'darkorange', '*': 'crimson'}
## 코드작성
marker_list=['o','^','*']
for m in df['marker'].unique():
    d1=df.loc[df['marker']==m]
    _=ax.scatter(d1['x'], d1['y'], marker=m, color=color_dic[m] ,edgecolor='k', s=300, label=color_dic[m])
    
_=ax.legend(bbox_to_anchor=(0.5,1), loc='lower center', ncol=3)
## 코드작성 끝
```

## 문제11

```python
# sr1을 bar graph로 나타내었다.
# 각 bar의 높이를 가지고 등수를 매기고,
# 그 등수를 bar에 결과그림과 같이 나타내시오

# 조건1) sr1의 값은 채점시 변경 가능


fig=plt.figure(figsize=(7,3), dpi=100)
ax1=fig.subplots()

sr1=pd.Series([7,4,2,5], index=range(4))
_=ax1.bar(sr1.index, sr1, color='y')
_=ax1.set_xticks(range(4))

#=== 코드작성 ===#
sr2=sr1.sort_values(ascending=False)

for x,y,rank in zip(sr2.index, sr2, range(1,5)):
    x,y,rank
    _=ax1.text(x,y,rank, ha='center', va='top', fontsize=16)
#=== 코드작성 끝 ===#
```

## 문제12

```python
# P(Px, Py)점은 랜덤하게 정해진다 (파란색 별).
# df에 있는 20개의 점들 또한 랜덤하게 정해진다 (노란색 원).
# df의 20개의 점들 중에서, P점에서 가장 멀리있는 점에 
# 1) 추가로 빨간색 별표를 scatter를 하고
# 2) 화살표로 가르키시오

# 힌트1) 추가 별표: color='r',s=50, marker='*'
# 힌트2) P점에서 점 x,y 까지의 거리: (x-Px)**2 + (y-Py)**2 (root를 생략해도 관계 없음)

fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

Px=np.random.randint(-2,3)
Py=np.random.randint(4,8)
_=ax.scatter(Px,Py, color='b',s=80, marker='*')

df=pd.DataFrame(np.random.uniform(-3,3,size=(20,2)), columns=['x','y'])
_=ax.scatter(df['x'],df['y'], color='y', edgecolor='k', s=150)

# #=== 코드작성 ===#
df['dst']=(df['x']-Px)**2 + (df['y']-Py)**2
df2=df.sort_values(by='dst')
df2
x,y,dst=df2.iloc[-1]
_=ax.scatter(x,y,color='r',s=50, marker='*')
_=ax.annotate('', (Px, Py), (x,y), arrowprops={'arrowstyle':'<-'})
# #=== 코드작성 끝 ===#
```
