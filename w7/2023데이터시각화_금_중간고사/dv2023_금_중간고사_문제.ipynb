## 문제7

```python
# 평균 2, 표준편차 2인 정규분포에서 3000개의 값을 뽑아서 히스토그램을 그렸다.
# 해당분포의 pdf 함수를 추가로 그리시오
# h1) plot line color='k'
# h2) pdf 함수의 x range: -4 < x < 8

fig=plt.figure(figsize=(5,4), dpi=100)
ax=fig.subplots()

_=ax.hist(np.random.normal(2,2,size=3000), bins=30, edgecolor='k', density=True)

## 코드작성
def pdf(X, mu, sigma):
    a=1/(sigma*np.sqrt(2*np.pi))
    b=-(X-mu)**2/(2*sigma**2)
    return a * np.exp(b)

X = np.linspace(-4,8,1000)
_=ax.plot(X,pdf(X,2,2), c='k')
```

## 문제8

```python
# 랜던하게 생성한 d1의 0번째 컬럼을 X, 1번째 컬럼을 Y로하여 scatter 하시오
# 단, x의 값을
# 3 나눈 나머지가 0이면 color_list의 0번째 color를
# 3 나눈 나머지가 1이면 color_list의 1번째 color를
# 3 나눈 나머지가 2이면 color_list의 2번째 color로 scatter 하시오
# c1) scatter 함수 한번만 호출
# c2) list comprehension 사용

fig=plt.figure(figsize=(5,3), dpi=100)
ax =fig.subplots()

d1=np.random.randint(6, size=(40,2))
color_list = ['r','k','b']

## 코드작성
colors = [color_list[val%3] for val in d1[:,0]]
_=ax.scatter(d1[:,0],d1[:,1],color=colors)
```

## 문제9

```python
# x1 < x < x2의 범위에서 11개의 값을 갖는 X가 있다. 함수 Y=X**2-2*X+3를 scatter하시오 (총 11개의 (x,y) 포인트를 scatter)
# c1) y값이 클수록 점이 진해지도록 코딩
# c2) 11개의 점이 희미하게라도 모두 보여야함
# c3) 채점시 x1, x2 값 변경 가능
# h1) scatter 함수의 s=100

x1=-4
x2=6

fig=plt.figure(figsize=(5,4), dpi=100)
ax=fig.subplots()

X=np.linspace(x1,x2,11)

## 코드작성
Y=X**2-2*X+3
min_Y = Y.min()
max_Y = Y.max()
alpha_list = (Y-min_Y+1)/(max_Y-min_Y+1)
_=ax.scatter(X,Y,alpha=alpha_list,s=100)
```

```python
# 주어진 df1은 30개의 행과 X,Y,M column으로 이루어져 있다.
# X,Y는 x,y의 좌표를 나타내고, M은 표현할 marker 모양을 나타낸다.
# x,y의 좌표에 해당 marker 모양으로 scatter 하시오
# c1) 점 하나 당 하나의 scatter 함수를 호출 (총 30번의 scatter 함수 호출)
# h1) scatter 함수 파라미터 값: color='y', edgecolor='k', s=80

fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

df1=pd.read_table('data/fri_p10.txt', sep='\t')

## 코드작성
for marker, data in df1.groupby('M'):
    _=ax.scatter(data['X'],data['Y'],marker=marker,color='y',edgecolors='k',s=80)
```

## 문제12

```python
# df는 x, y, price, station column을 가지고 있다.
# station에는 ['o','^','*'] 이 세 개의 값 중 하나가 들어있다.
# 아래의 조건을 참고하여, x,y 좌표에 마커를 scatter 하시오

# c1) 각 마커의 크기: price 값의 제곱
# c2) 각 마커의 모양: station에 있는 값
# c3) scatter 함수는 3번 (등장하는 마커의 개수)만 호출
# h1) color='y',edgecolor='k'


fig=plt.figure(figsize=(5,5), dpi=100)
ax=fig.subplots()

df = pd.read_table('data/fri_p12.txt', sep='\t')
df.head()

## 코드작성
for marker, data in df.groupby('station'):
    _=ax.scatter(data['x'],data['y'],marker=marker, s=data['price']**2, color='y', edgecolors='k')
```
