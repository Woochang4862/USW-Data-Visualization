## 문제2

```python
# tue_p2.txt 파일을 읽은 data가 있다.
# 세 가지 색깔의 해당 column들이, col_red, col_green, col_blue에 담겨있다.
# 각 column을 해당 color로 plot하시오

# c1) for loop을 사용
# c2) col_red, col_green, col_blue의 column들은 채점시 변경 가능

fig=plt.figure(figsize=(6,4), dpi=100)
ax=fig.subplots()

data=pd.read_table('data/tue_p2.txt', sep='\t')
data.head()

col_red=['Y1','Y4','Y7']
col_green=['Y2','Y5', 'Y8']
col_blue=['Y3','Y6', 'Y9', 'Y10']
color_column_dic = {'r':col_red,'g':col_green,'b':col_blue}

## 코드작성
for color in ['r','g','b']:
    for col in color_column_dic[color]:
        _=ax.plot(data[col],color=color)
```
