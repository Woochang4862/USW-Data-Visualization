# bar

## bar 예제 1

```python
fig=plt.figure(figsize=(10,5), dpi=100)
axs=fig.subplots(1,2)

data = np.array([5,25,50,20])
axs[0].bar(range(len(data)),data)
axs[1].bar(range(len(data)),data, align='edge')
```

## bar: various width

```python
fig=plt.figure(figsize=(12,5), dpi=100)
axs=fig.subplots(1,3)

def bar_width(ax,data,w):
    ax.bar(range(len(data)),data, width=w)
    ax.set_title('width: {}'.format(w))
    
data = np.array([5,25,50,20])
bar_width(axs[0], data, w=0.8) # default width: 0.8
bar_width(axs[1], data, w=0.5)
bar_width(axs[2], data, w=1)
```

## Series bar plot

```python
fig=plt.figure(figsize=(13,3), dpi=100)
axs=fig.subplots(1,3)

sr1=pd.Series(data=[10,9,7,8], index=['a1','a2','a3','a4'])
sr1
axs[0].bar(sr1.index, sr1)
axs[1].bar(sr1.index, sr1.values)
sr1.plot.bar(ax=axs[2])
```

## 다중 막대차트 그리기

```python
fig=plt.figure(figsize=(8,3), dpi=100)
ax1,ax2=fig.subplots(1,2)

data=np.array([[5,25,50,20],
               [4,23,51,17],
               [6,22,52,19]])

X=np.arange(4)

ax1.bar(X+0.00, data[0], width=0.25)
ax1.bar(X+0.25, data[1], width=0.25)
ax1.bar(X+0.50, data[2], width=0.25)

## ax2
for ii in np.arange(data.shape[0]):
    _=ax2.bar(X+ii*0.25, data[ii], width=0.25)
```

## 분할 막대차트 그리기

```python
fig=plt.figure(figsize=(10,5), dpi=100)
ax1,ax2=fig.subplots(1,2)

data=np.array([[20,10,35,20],
               [10,12,9,10],
               [5,6,6,5]])
X=np.arange(4)

_=ax1.bar(X,data[0])
_=ax1.bar(X,data[1],bottom=data[0])
_=ax1.bar(X,data[2],bottom=data[0]+data[1])

for i in range(len(data)):
    _=ax2.bar(X, data[i], bottom=np.sum(data[:i], axis=0))
```

```python
fig=plt.figure(figsize=(8,5), dpi=100)
ax1, ax2=fig.subplots(1,2)

data=pd.read_table('data/dat_bar.txt',sep='\t', index_col=0)
data

## ax1
color_list=['r','g','b','y','c','m','r','g','b','y','c','m']
X=np.array([0,1])
for i, color in enumerate(color_list):
    _=ax1.bar(X,data.iloc[i],bottom=np.sum(data.iloc[:i], axis=0),color=color)
ax1.set_xticks(X,labels=data.columns)

## ax2
_=data.T.plot.bar(ax=ax2,stacked=True)
```
