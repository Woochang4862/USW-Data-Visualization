```python
fig=plt.figure(figsize=(15,4), dpi=150)
ax1, ax2, ax3=fig.subplots(1,3)

############## ax1
# d1, d2가 주어져 있다.
# d1을 boxplot으로 x=0, x=10에 나타내고, 
# d2를 histogram으로 나타내시오 (histogram parameter: bins=50, edgecolor='k')

d1=np.random.normal(loc=10, scale=2, size=500)
d2=np.random.normal(loc=5, scale=1, size=500)

## 코드작성
_=ax1.boxplot([d1,d1],positions=[0,10])
_=ax1.hist(d2, bins=50, edgecolor='k')

############## ax2
# uniform distribution (2,5)에서 뽑은 100개 값의 '최소값'을 m1이라고 할 때,
# 서로 다른 m1의 값 1000개를 구하여 히스토그램으로 ax2에 나타내시오
# - 평균이 아닌 최소값에 주의 (최대값구하는 함수: min)
# - bins=20, edgecolor='k'

## 코드작성
m1s = np.min(np.random.uniform(2,5,size=(1000,100)), axis=1)
print(f"100개의 원소를 가진 집합 m1의 최대 원소 개수 : {max(np.unique(m1s,return_counts=True)[1])}")
_=ax2.hist(m1s, bins=20,edgecolor='k')

############## ax3
# 랜던하게 생성한 d1의 0번째 컬럼을 X, 1번째 컬럼을 Y로하여 scatter 하시오
# 단, x의 값을
# 3 나눈 나머지가 0이면 color_list의 0번째 color를
# 3 나눈 나머지가 1이면 color_list의 1번째 color를
# 3 나눈 나머지가 2이면 color_list의 2번째 color로 scatter 하시오
# c1) scatter 함수 한번만 호출
# c2) list comprehension 사용

d1=np.random.randint(6, size=(40,2))
color_list = ['r','k','b']
X, Y = d1[:,0], d1[:,1]
colors = [color_list[x%3] for x in X]
_=ax3.scatter(X,Y, c=colors)

## 코드작성
```
