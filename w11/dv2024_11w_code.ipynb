### 실습6

```python
fig=plt.figure(figsize=(5,3), dpi=150)
ax=fig.subplots()

def pdf(X, mu, sigma):
    a = 1/(sigma * np.sqrt(2*np.pi))
    b = -1/(2*(sigma**2))
    return a * np.exp(b * ((X - mu)** 2))

m1=1
s1=2

X = np.linspace(m1-5, m1+5, 1000)
_=ax.plot(X, pdf(X, m1, s1), color = 'k')

_=ax.set_xticks([m1],labels=['$\\mu$={}'.format(m1)], fontsize=16)

## 코드작성
_=ax.plot([m1-2,m1+2],[pdf(m1-2,m1,s1),pdf(m1+2,m1,s1)])
_=ax.text(m1,pdf(m1-2,m1,s1),f'$\\sigma={s1}$', ha='center', fontsize=12, va='bottom')
```

## legend 사용 두번째 방법 (심화 사용)

### legend의 parameter로 label을 사용하는 경우 (plot)

```python
fig=plt.figure(figsize=(8,3), dpi=150)
ax1, ax2=fig.subplots(1,2)

X=np.linspace(0,2*np.pi, 100)
Y1=np.sin(X)

## ax1 (method1)
_=ax1.plot(X,Y1, c='k', lw=3, ls='--', label='sin(x)')
_=ax1.legend()

## ax2 (method2)
l1=ax2.plot(X,Y1, c='k', lw=3, ls='--')
l1
l1[0]
_=ax2.legend(handles=(l1[0],), labels=('sin(x)',))
```

### legend의 parameter로 label을 사용하는 경우 (plot, scatter)

```python
fig=plt.figure(figsize=(13,4), dpi=100)
ax1, ax2=fig.subplots(1,2)

X=np.linspace(0,2*np.pi, 100)
Y1=np.sin(X)
Y2=np.cos(X)
Y3=np.random.uniform(-1,1,size=100)

## ax1
_=ax1.plot(X,Y1, c='k', lw=3, ls='--',label='sin(x)')
_=ax1.plot(X,Y2, c='r', lw=1, ls=':',label='cos(x)')
_=ax1.scatter(X,Y3, marker='*', label='Star')
_=ax1.legend(title='plot & scatter',fontsize=13, title_fontsize=16)

## ax2
plot1=ax2.plot(X,Y1, c='k', lw=3, ls='--')
plot2=ax2.plot(X,Y2, c='r', lw=1, ls=':')
scat1=ax2.scatter(X,np.random.uniform(-1,1,100), marker='*')
plot2[0]
scat1

_=ax2.legend(handles=(plot1[0], plot2[0], scat1), labels=('sin(x)', 'cos(x)', 'Star'),
             title='plot & scatter',fontsize=13, title_fontsize=16)
```
